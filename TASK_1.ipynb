Task 1: Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('climate-ds.csv',index_col=0)
df.head()
```

```python
#rows and columns
df.shape
```

```python
#information about data
# check data type of each coulmn
df.info()
```

```python
#some statistical information about data
df.describe()
```

```python
def hist_box():
    for i in df.columns[3:]:
        plt.subplots(nrows=1,ncols=2,figsize=(15,6))
        plt.subplot(1,2,1)
        plt.hist(df[i])
        plt.title(i);
        plt.subplot(1,2,2)
        sns.boxplot(data=df[i])
        plt.title(i);
        yield()
```

```python
hibox = hist_box()
next(hibox);
```
